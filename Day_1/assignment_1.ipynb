```python
import numpy as np
import pandas as pd
```

```python
class dframe:
    df = pd.DataFrame()
    def generateRandomN(self, ros, cols):
        df = pd.DataFrame(data=np.random.randn(ros,cols))
        return df
        
    def convertToDataFrame(self, df):
        return df[5].plot.hist()
```

```python
fm = dframe()
user_df = fm.generateRandomN(10000,10)
user_df.to_csv("user_dataframe.csv")
fm.convertToDataFrame(user_df)
```

#### Observation
as per the above histogram of column 5, value between -1 to 0 is appear more than 2800 times which is the highest number the lowest one is value between 3 to 4.
